1.Tokenizacion, Lematizacion y Stemming



```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer, SnowballStemmer

#Descarga de recursos
nltk.download('punkt')
nltk.download('wordnet')

#Texto a tokenizar
text = "Michael Jackson is singing a new song."

print("Tokenizacion: ")
print(sent_tokenize(text))

print("\nWords: ")
print(word_tokenize(text))

#NLP
lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer("english")

#Palabras a analizar
words = ["singing", "sang","sings"]

#Resultados
print("\n{:<12}{:<12}{:<12}".format("Word","Lemmatizer","Stemmer"))
for word in words:
  lemma = lemmatizer.lemmatize(word, pos='v')
  stem = stemmer.stem(word)

  print("\n{:<12}{:<12}{:<12}".format(word, lemma,stem))
```
